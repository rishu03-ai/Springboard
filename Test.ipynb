```python
# YOLOv11n inference on test images, saving annotated results in runs/detect/test
from ultralytics import YOLO
import os
from glob import glob
import cv2

# Path to trained weights and test images
model_path = r"runs/detect/train/weights/best.pt"
test_img_dir = r"MyFirstProject/test/images"
output_dir = r"runs/detect/test"
os.makedirs(output_dir, exist_ok=True)

# Load trained model
model = YOLO(model_path)

# Run inference and save annotated images in output_dir
test_imgs = glob(os.path.join(test_img_dir, "*.jpg"))
for img_path in test_imgs:
    results = model.predict(source=img_path, conf=0.25, save=False)
    annotated_img = results[0].plot()
    base_name = os.path.basename(img_path)
    save_path = os.path.join(output_dir, base_name)
    cv2.imwrite(save_path, annotated_img)
```
