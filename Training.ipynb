```python
from ultralytics import YOLO

# 1. Load the pre-trained YOLOv11 nano model
model = YOLO('yolo11n.pt') 

# 2. Train the model for only 5 epochs (for a quick test)
# Replace 'data.yaml' with your actual dataset config file name if different.
model.train(
    data='MyFirstProject/data.yaml', 
    epochs=50,          # Minimal epochs for a quick run
    imgsz=512,         # Standard input size
    batch=16           # Standard batch size
)

# 3. Evaluation (Validation)
# The best model weights will be saved automatically, usually under runs/detect/trainX/weights/best.pt
# This command automatically validates the final trained model on the validation set.
metrics = model.val() 

# 4. Print the key result
print(f"\nTraining and Validation Complete.")
print(f"Mean Average Precision (mAP@0.5:0.95): {metrics.box.map:.4f}")
```
